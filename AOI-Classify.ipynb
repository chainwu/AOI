```python
import keras
import matplotlib
import imageio
import matplotlib.pyplot as plt
import numpy as np
import glob
import os.path as path

from keras.models import load_model

AOI = 'aoi.h5'
images=[]
test_predictions = None
img_rows = 128
img_cols = 128

print("----")

model = load_model(AOI)
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Image processing
IMAGE_PATH='test_images'
image_input_files = sorted(glob.glob(path.join(IMAGE_PATH, '*.png')))

images = [imageio.imread(im) for im in image_input_files]
images = np.asarray(images)
#images = images[..., None]
n_images = len(image_input_files)
images = images
images = images.reshape(images.shape[0], img_rows, img_cols, 1)
test_predictions = model.predict_classes(images, verbose=1)
print(test_predictions)
```

```python
for i in range(len(test_predictions)):
    print("test_predition", i, ":", test_predictions[i])
```
