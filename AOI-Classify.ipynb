```python
import keras
import matplotlib
import imageio
import matplotlib.pyplot as plt
import numpy as np
import glob
import os.path as path
import tensorflow as tf
from keras.models import load_model

AOI = 'aoi-cnn.h5'

print("----")

model = load_model(AOI)
model.summary()
```

```python
tf.print(model.layers[0].output)
```

```python
IMAGE_PATH='test_images/'
num_classes = 6
image_rows = 128
image_cols = 128
    
for i in range(df.shape[0]):
    image_input_file = IMAGE_PATH + df.ID[i]
    #print(image_input_file)
    orig_image = imageio.imread(image_input_file)
    image = np.asarray(orig_image)
    image = image/256
    #print(image)
    image=image.reshape(1, image_rows, image_cols, 1)
    #x_train = x_train.reshape(x_train.shape[0], img_rows * img_cols, 1)
    img_class = model.predict_classes(image)
    prediction = img_class[0]
    df.Label[i]=img_class[0]
    print("Image ",i,":", prediction)
    
    #plt.axis('off')
    #plt.imshow(orig_image)
    #title = "Image "+str(i)+":"+str(img_class[0])
    #plt.title(title)
    #plt.show()
```
